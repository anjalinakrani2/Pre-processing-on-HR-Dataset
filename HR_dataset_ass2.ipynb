# **Apply Binning (discretize continuous value ) to salary attribute**

## **Bin by mean**

```python
salary=df['Salary']
#bin_means=pd.cut(salary,bins=3,labels=False)
bin_means=pd.cut(salary,bins=3,labels=False)
```

```python
bin_means
```

```python
df['salary_bin_means']=salary.groupby(bin_means).transform('mean')
df['salary_bin_means'].head()
```

```python
df['salary_bin_means'].unique()
```

## **Bin by boundaries**

```python
df['Salary'].min()
```

```python
df['Salary'].max()
```

```python
boundaries=[0,60000,120000,180000,np.inf]
bin_label=['Low','Medium','High','Very High']
df['salary_bin_boundary']=pd.cut(salary,bins=boundaries,labels=bin_label)
```

```python
df[['Salary', 'salary_bin_means', 'salary_bin_boundary']].head()
```

# **Apply Data transformation on salary attribute**

## **Min-max normalization**

```python
min_salary=salary.min()
max_salary=salary.max()
```

```python
df['salary_MinMax']=(salary-min_salary)/(max_salary-min_salary)
```

```python
df[['Salary','salary_MinMax']].head()
```

## **Z-score normalization**

```python
mean_salary=salary.mean()
std_salary=salary.std()
```

```python
df['salary_zScore']=(salary-mean_salary)/std_salary
```

```python
df[['Salary','salary_zScore']].head()
```

## **Decimal scaling**

```python
#ceil(log10(x)) will give number of digits of x
j = np.ceil(np.log10(salary.abs().max()))
```

```python
df['salary_decimal_scaling']=salary/10**j
```

```python
df[['Salary','salary_decimal_scaling']].head()
```
